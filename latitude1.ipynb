```python
import pandas as pd
```

```python
df_1=pd.read_csv("places.csv")
```

```python
df_1.head()
```

```python
df_1['lat']=""
df_1["lng"]=""
df_1["class"]=""
```

```python
import requests
import urllib.parse
from tqdm import tqdm_notebook as tqdm
wrong_l=[]
lat=[]
long=[]
for i in tqdm(range(len(df_1))):
    try:
        address = df_1['Place'][i]
    #         print("Step1")
        url = 'https://nominatim.openstreetmap.org/search/' + urllib.parse.quote(address) +'?format=json'
    #         print("Step2")
        response = requests.get(url).json()
        print(response)
        df_1["class"][i]=(response[0]['class'])
    #         print(response[0]['lat'])
    #         print("Step3")
        df_1['lat'][i]=(response[0]['lat'])
# #         print("Step4")
        df_1['lng'][i]=(response[0]['lon'])  
          
#         print("Step5")
    except KeyError:
        wrong_l.append(i)
    except IndexError:
        wrong_l.append(i)
    continue
```
